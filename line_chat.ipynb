# Import Required Libraries

Import the necessary libraries, including pandas and matplotlib.


```python
# Import Required Libraries
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the Dataset

Load the dataset using pandas' read_csv function.


```python
# Load the Dataset
df_canada = pd.read_csv('canadian_immegration_data.csv')
```

# Prepare the Data

Select the data for Haiti and the relevant years.


```python
# Prepare the Data
years = list(map(str, range(1980, 2014)))  # Convert years to string
haiti_data = df_canada.loc[df_canada['Country'] == 'Haiti', years].transpose()  # Select Haiti data and transpose
haiti_data.index = haiti_data.index.map(int)  # Convert index to integer for plotting
haiti_data.columns = ['Number of Immigrants']  # Rename the column
haiti_data.reset_index(inplace=True)  # Reset index to make 'Year' a column
haiti_data.rename(columns={'index': 'Year'}, inplace=True)  # Rename 'index' to 'Year'
```

# Create the Line Chart

Use matplotlib to create and display the line chart.


```python
# Create the Line Chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(haiti_data['Year'], haiti_data['Number of Immigrants'], marker='o')  # Plot the data with markers
plt.title('Immigration from Haiti')  # Set the title of the plot
plt.ylabel('Number of Immigrants')  # Set the y-axis label
plt.xlabel('Years')  # Set the x-axis label
plt.grid(True)  # Enable the grid
plt.show()  # Display the plot
```
